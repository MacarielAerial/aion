# Image Resizing

## Proof of Concept - No Deep Learning

### Import Python Dependencies

```python
import pandas as pd
import numpy as np
from PIL import Image
from IPython.display import display
```

### Load Input Data

```python
df = pd.read_csv('../data/01_raw/Challenge2.csv').dropna()
df.tail(22)
```

### Preprocess Raw Input into PIL / OpenCV Compatible Object

```python
# Group rows together by frames
def group_rows_by_depth(df, rows_per_frame):
    grouped_frames = []
    num_frames = len(df) // rows_per_frame
    for i in range(num_frames):
        start = i * rows_per_frame
        end = start + rows_per_frame
        frame = df.iloc[start:end].drop(columns=['depth']).values
        grouped_frames.append(frame)
    return grouped_frames
rows_per_frame = 10
grouped_frames = group_rows_by_depth(df, rows_per_frame)


# Preprocess frames
preprocessed_frames = []
for frame_data in grouped_frames:
    combined_frame = np.vstack(frame_data)
    normalized_frame = (combined_frame / 255.0) * 255.0
    preprocessed_frames.append(normalized_frame.astype(np.uint8))

# Check one frame
sample_frame = preprocessed_frames[-50]
sample_image = Image.fromarray(sample_frame)
display(sample_image)
```

### Resize Frames

```python
# Define the function to resize images
def resize_image(data, new_width):
    height, _ = data.shape
    image = Image.fromarray(data)
    resized_image = image.resize((new_width, height), Image.LANCZOS)
    return np.array(resized_image)

# Resize frames
processed_frames = []
new_width = 150
for frame in preprocessed_frames:
    resized_frame = resize_image(frame, new_width)
    processed_frames.append(resized_frame)

# Check one frame
sample_frame = processed_frames[-50]
sample_image = Image.fromarray(sample_frame)
display(sample_image)
```

### Query Frames Based on depth_min and depth_max

```python
def request_frames(depth_min, depth_max, depths, frames, rows_per_frame):
    filtered_frames = []
    filtered_depths = []

    # Calculate frame boundaries
    depth_min_frame = int(depth_min)
    depth_max_frame = int(depth_max)

    # Iterate over frames and check if they fall within the depth range
    for i in range(len(frames)):
        frame_depth_start = int(depths[i * rows_per_frame])
        frame_depth_end = int(depths[(i + 1) * rows_per_frame - 1])
        
        if frame_depth_start >= depth_min_frame and frame_depth_end <= depth_max_frame:
            filtered_frames.append(frames[i])
            filtered_depths.append((frame_depth_start, frame_depth_end))
    
    return filtered_depths, filtered_frames

# Example usage
depth_min = 9100.5
depth_max = 9353.9

depths = df['depth'].tolist()
filtered_depths, filtered_frames = request_frames(depth_min, depth_max, depths, processed_frames, rows_per_frame)

# Convert and display a sample frame
if filtered_frames:
    sample_frame = filtered_frames[-50]
    sample_image = Image.fromarray(sample_frame)
    display(sample_image)
else:
    print("No frames found within the specified depth range.")
```

### Apply Colour Map to Frames

```python
import cv2

# Define a function to apply colour map to frames
def apply_colour_map(frames, color_map=cv2.COLORMAP_JET):
    colored_frames = []
    
    for frame in frames:
        # Combine the rows of the frame into a single image array
        combined_frame = np.vstack(frame).astype(np.uint8)
        
        # Apply the color map using OpenCV
        coloured_frame = cv2.applyColorMap(combined_frame, color_map)
        
        # Convert the colored frame to a PIL Image
        # image = Image.fromarray(cv2.cvtColor(colored_frame, cv2.COLOR_BGR2RGB))
        colored_frames.append(coloured_frame)
    
    return colored_frames

coloured_frames = apply_colour_map(filtered_frames, color_map=cv2.COLORMAP_VIRIDIS)

# Visualise one frame
sample_frame = coloured_frames[-20]
sample_image = Image.fromarray(sample_frame)
display(sample_image)
```
